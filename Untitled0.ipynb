```python
from pandasql import sqldf
```

```python
query = """
SELECT
    driver_id,
    AVG(driver_rating) AS avg_rating,
    COUNT(order_id) AS total_orders
FROM driver_performance
GROUP BY driver_id
ORDER BY avg_rating DESC
LIMIT 10
"""

sqldf(query)
```

```python
query2 = """
SELECT
    customer_id,
    loyalty_score,
    COUNT(order_id) AS total_orders
FROM orders
JOIN customers USING (customer_id)
GROUP BY customer_id
ORDER BY loyalty_score DESC
LIMIT 10
"""

sqldf(query2)
```

```python
query3 = """
SELECT
    pickup_zone,
    SUM(compensation_amount) AS total_compensation
FROM complaint_delivery
GROUP BY pickup_zone
ORDER BY total_compensation DESC
"""

sqldf(query3)
```

```python
query4 = """
SELECT
    success_flag,
    AVG(api_latency_ms) AS avg_latency
FROM app_events
GROUP BY success_flag
"""

sqldf(query4)
```

```python
query5 = """
SELECT
    severity,
    AVG(resolved_hours) AS avg_resolution_hours,
    COUNT(*) AS incident_count
FROM incidents
GROUP BY severity
ORDER BY avg_resolution_hours DESC
"""
sqldf(query5)
```

```python
query6 = """
SELECT
    booking_channel,
    COUNT(CASE WHEN delivery_status = 'Success' THEN 1 END) * 100.0 / COUNT(*) AS success_rate_pct
FROM delivery_orders
GROUP BY booking_channel
ORDER BY success_rate_pct DESC
"""
sqldf(query6)
```
